# Normalizing and Scaling data


```python
from sklearn.preprocessing import StandardScaler # importing StandardScaler from sklearn
std_scale = StandardScaler()
std_scale 
```

Scaling only the numerical variables as few have large numbers like salery and some have very small number i=limited to 1 or 2 digits. 

```python
X['Bonus'] = std_scale.fit_transform(X[['Bonus']])


X['low'] = std_scale.fit_transform(X[['low']])


X['closingbalance'] = std_scale.fit_transform(X[['closingbalance']])
X['Age'] = std_scale.fit_transform(X[['Age']])
```

## Let us now split the data into TEST and TRAIN

```python
from sklearn.model_selection import train_test_split

# Split the data into train and test with 80% data in train and Random state as 5.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=5)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
y_train_predict = lin_reg.predict(X_train)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
r2 = r2_score(y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

```python
y_test_predict = lin_reg.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))
r2 = r2_score(y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

The RMSE (roor mean square error )is e-06 order, which is very low and the r2_score is also 1, which means that the selected features work well with the regression model and The R^2 value is 1.0 meaning our model is able to explain 100% variability in the data.

## The low RMSE value and perfect R squared value for both the training and testing data suggests a good prediction model.
